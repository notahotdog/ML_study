```python
# reference https://towardsdatascience.com/knn-visualization-in-just-13-lines-of-code-32820d72c6b6
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, neighbors
from mlxtend.plotting import plot_decision_regions 
# For almost anything that you want to do by Python, there are more than one way of coding.
# mlxtend is one popular package for basic data mining, 
# if this is the first time you use mlxtend, you need to install it first
# for novice, you need to use cmd in Windows, and type pip install mlxtend 

# Core KNN function
# This version is not a typical one, we transform data to create X and Y and then build a classifier
# later, you will see more typical flow of codes many times throughout the semester.
def knn_comparison(data, k):
 # Features which contribute to knn
 x = data[["X","Y"]].values # x,y co-ordinates of a point - # [[0.03,1.1],[],..[]] - Nested tuple
#  print(f"x: {x}") 
 # Classification associated with the datapoint
 y = data["class"].astype(int).values # Each point by default already gets assigned a class (in this example it is either 1 or 0)
#  print(y)
 clf = neighbors.KNeighborsClassifier(n_neighbors=k) # key step to specify model, you will use many times this semester

 # Pass in the features, and their associated classification for training
 clf.fit(x, y) # key step to train a model on your training dataset

 # Plotting decision region by a unique function plot_decision_regions()
 plot_decision_regions(x, y, clf=clf, legend=2)
 
# Adding axes annotations
 plt.xlabel("X")
 plt.ylabel("Y")
 plt.title("Knn with K="+ str(k))
 plt.show()
```

```python
# other datasets for training by KNN
# Example 1:True Pattern is two U-Shapes
data1 = pd.read_csv("ushape.csv") # now we are loading input data from a file
for i in [1,5,20,30,40]:
    knn_comparison(data1, i)
```

From the observations above, we can see the difference for when there is a low k value set - the model over fit's
and vv when there is a high k value - it under fit's
